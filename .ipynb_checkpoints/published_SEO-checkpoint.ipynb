Using the Google API (together with the gspread module) to  access google spreadsheets. In this exercise we are specifically accessing updates in SEO-ready pages being produced and published in different websites throughout time.

```python
import gspread as gs
import pandas as pd

#get my google API
gc = gs.service_account(filename='seo-sites-418012-45dcf5614e99.json')

#open the google spreadsheet file
sh = gc.open_by_url('https://docs.google.com/spreadsheets/d/14E5iVTB8J8EVIlTPMkorhEGYr4S3ggLZ0MTgnoF7gpY/edit#gid=0')
```

Because there are multiple tabs in my spreadsheet, each corresponding to one website, I'll import them separately, clean each dataframe and then concatenate all in one dataframe

MS website:

```python
#lets me know how many worksheets are in my googledoc
worksheet_list = sh.worksheets()

#select the name of the worksheet I want
ws = sh.worksheet('MS 1')

df = pd.DataFrame(ws.get_all_values())

#drop the first row, should not have the name of the values
df1 = df.drop(0)

#create new column...
df1['MS'] = 'MS'

#...and set it as index (this will be for concatenating and group by disease later on)
df1.set_index('MS',inplace=True)

#rename my columns
df2 = df1.rename(columns ={1: 'name', 5: 'status'})

#keep only the columns I want
df_ms = df2[['name','status']]

df_ms.head()
```

Now, to concatenate all dataframes and itirate through the different rows:

```python
#select the name of the worksheet I want
ws8 = sh.worksheet('FAB')

df = pd.DataFrame(ws8.get_all_values())

#drop the first row, should not have the name of the values
df1 = df.drop(0)

#create new column...
df1['FAB'] = 'FAB'

#...and set it as index (this will be for concatenating and group by disease later on)
df1.set_index('FAB',inplace=True)

#rename my columns
df2 = df1.rename(columns ={1: 'name', 5: 'status'})

#keep only the columns I want
df_fab = df2[['name','status']]
df_fab.head()
```

```python
#join dataframes
frames = [df_ms, df_als, df_pomp, df_sma, df_lup, df_fa, df_park, df_fab]

big_df = pd.concat(frames)

#iterate through dataframes and keep values for published pages in each worksheet
published = [] 

for label, row in big_df.iterrows():
    if row['status'] == 'Published' and row.name == 'MS':
        published.append('ms')
    elif row['status'] == 'Published' and row.name == 'ALS':
        published.append('als')
    elif row['status'] == 'Published' and row.name == 'POMP':
        published.append('pomp')
    elif row['status'] == 'Published' and row.name == 'SMA':
        published.append('sma')
    elif row['status'] == 'Published' and row.name == 'LUP':
        published.append('lup')
    elif row['status'] == 'Published' and row.name == 'FA':
        published.append('fa')
    elif row['status'] == 'Published' and row.name == 'PARK':
        published.append('park')
    elif row['status'] == 'Published' and row.name == 'FAB':
        published.append('fab')
```

Finally, plot the histogram. The advantage of this process is that as different SEO pages are updated and published in each website, we can have an idea in real time of the changes happening.

```python
import matplotlib.pyplot as plt
from aquarel import load_theme
import seaborn as sns

# Define a custom color palette
custom_palette = sns.color_palette("husl", 12)  # Example palette with 12 colors


theme = load_theme("arctic_dark")
theme.apply()

# Data
labels = ['MS', 'ALS', 'POMP', 'SMA', 'LUP', 'FA', 'PARK', 'FAB']

counts = [published.count('ms'), published.count('als'), published.count('pomp'),
          published.count('sma'), published.count('lup'), published.count('fa'),
          published.count('park'), published.count('fab')
         ]

# Create the histogram
plt.bar(labels, counts, color= custom_palette)

# Add labels and title
plt.xlabel('Sites')
plt.ylabel('Counts')
plt.title('Histogram of Published SEO pages per Site')

# Add value annotations on top of each bar
for i, count in enumerate(counts):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

#apply aquarel theme
theme.apply_transforms()

# Show the plot
plt.show()
```
